```python
# You can filter with the python comparison operators: == < >. 

# Where did the Greens get more votes than the Conservatives

parties[parties["green"] > parties["con"]]
```

```python
# You can also use & (and) and | (or) to perform logical queries. 
# Just bracket your queries and use a & or |

# Where did the Greens get more votes than the Conservatives OR Labour?

parties[ ( parties["green"] > parties["con"] ) | ( parties["green"] > parties["lab"] ) ]
```
